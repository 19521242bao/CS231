```python
def ShowPredict(imagepath):
  import cv2
  from google.colab.patches import cv2_imshow
  image = cv2.imread(imagepath)

  colors = {
      'No entry':(178, 0, 31),
      'No parking':(241, 175,	0),
      'No turning':(91, 189, 43),
      'Max speed':(0, 166, 173),
      'Other prohibition signs':(81, 31, 144),
      'Warnings': (162, 0, 124),
      'Mandatory':(249,244,0)
  }

  f = open('result.txt', 'r')
  count = 0
  annotation = []
  for line in f:
    if (count >= 13):
      annotation.append(line.replace('\n', ''))
    count += 1
  
  for anno in annotation:
    class_ = anno[:anno.find(':')]
    predict = anno[anno.find(':')+2:anno.find('(')].replace('\t', '')
    predict = int(predict[:-1])
    tmp = anno[anno.find('('):].split()
    (x_min, y_min) = int(tmp[1]), int(tmp[3])
    (x_max, y_max) = int(tmp[5]) + x_min, int(tmp[7][:-1]) + y_min

    cv2.rectangle(image, (x_min, y_min), (x_max, y_max), colors[class_], 3)
    cv2.putText(image, "{} [{:.2f}]".format(class_, predict),
                (x_min, y_min - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5,
                colors[class_], 3)
  cv2_imshow(image)
  cv2.imwrite('predictions.jpg', image)

ShowPredict('predictions.jpg')
```

```python
import os
from PIL import Image
def get_all_bboxes(annot_path,img_path):
    img = Image.open(img_path)
    width, height = img.size
    list_info = []
    with open(annot_path) as f:
      content = f.readlines()
    
    content = [x.strip() for x in content] 
    for line in content:
      line = [float(i) for i in line.split()]
      class_id = int(line[0])
      # Chuyn nh dng class x_center y_center w h v nh dng xmin,xmax,ymin,ymax.
      x_min=int((line[1] - line[3]/2) * width)
      y_min=int((line[2] - line[4]/2) * height)
      x_max=int((line[1] + line[3]/2) * width)
      y_max=int((line[2] + line[4]/2) * height)
      bbox = [(x_min), (y_min), (x_max), (y_max)]
      info_box = [class_id, bbox]
      list_info.append(info_box)
  
    return list_info
```

```python
def draw_image(image_name):
  image = cv2.imread(image_name)
  list_info=get_all_bboxes(annot_path,image_name)
  for info in list_info:
    class_id = info[0]
    bbox = info[1]
    # draw rectangle
    image = cv2.rectangle(image, (bbox[0], bbox[1]), (bbox[2], bbox[3]), (255, 255, 0), 2)
    # write class_id
    image = cv2.putText(image, str(class_id), (bbox[0], bbox[1]-3), cv2.FONT_HERSHEY_SIMPLEX , 1, (255, 0, 0), 2, cv2.LINE_AA)
  
  return image
```
