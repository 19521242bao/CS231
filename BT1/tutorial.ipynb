# Loading and displaying an image

```python
import imutils
import cv2
import matplotlib.pyplot as plt
# import necessary library
```

```python
import cv2
from matplotlib import pyplot as plt

def showImage(image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.show()
```

```python
#load image
image = cv2.imread("Denis_Mukwege.jpg")
(h, w, d) = image.shape
# h,w,d with height,weight,dimension picture
print("width={}, height={}, depth={}".format(w, h, d))
# show imgae
# font = cv2.FONT_HERSHEY_SIMPLEX
# # org
# org = (50, 50) 
# # fontScale
# fontScale = 1
# # Blue color in BGR
# color = (255, 0, 0)
# cv2.putText(image,"hello",org,font,fontScale,color)
cv2.imshow("Image", image)
#showImage(image)
cv2.waitKey(0)
# dùng để dừng quá trình display tấm ảnh bằng việc nhấn phím bất kì
```

# Accessing individual pixels

```python
# truy cập RGB pixel x=50, y=100 với x là chiều
#x:weight
#y:height
# OpenCV stores images in BGR order rather than RGB
(B, G, R) = image[100, 80]
#height is the number of rows and the width is the number of columns
print("R={}, G={}, B={}".format(R, G, B))
```

# Array slicing and cropping

```python
# extract a 100x100 pixel square ROI (Region of Interest) from the
# input image starting at x=150,y=70 at ending at x=600,y=720
roi = image[70:720,150:600]# face size
showImage(roi)
# cv2.imshow("Roi",roi)
# cv2.waitKey(0)
#
```

# Resize Image

```python
# resize the image to 200x200px, ignoring aspect ratio
resized = cv2.resize(image, (200, 200))
showImage(resized)
# cv2.imshow("Fixed Resizing", resized)
# cv2.waitKey(0)
```

```python
r = 300.0 / w
dim = (300, int(h * r))# kích thước điều chỉnh phù hợp về kích thước 300xdim
resized = cv2.resize(image, dim)
showImage(resized)
# cv2.imshow("Aspect Ratio Resize", resized)
# cv2.waitKey(0)
```

```python
# manually computing the aspect ratio can be a pain so let's use the
# imutils library instead
resized = imutils.resize(image, width=300)
# cv2.imshow("Imutils Resize", resized)
# cv2.waitKey(0)
showImage(resized)
```

# Rotating an image

```python
# let's rotate an image 45 degrees clockwise using OpenCV by first
# computing the image center, then constructing the rotation matrix,
# and then finally applying the affine warp
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, -45, 1.0)
rotated = cv2.warpAffine(image, M, (w, h))
# cv2.imshow("OpenCV Rotation", rotated)
# cv2.waitKey(0)
showImage(rotated)
```

```python
rotated = imutils.rotate(image, -45)
# cv2.imshow("Imutils Rotation", rotated)
# cv2.waitKey(0)
showImage(rotated)
```

```python
# OpenCV doesn't "care" if our rotated image is clipped after rotation
# so we can instead use another imutils convenience function to help
# us out
rotated = imutils.rotate_bound(image, 90)
# cv2.imshow("Imutils Bound Rotation", rotated)
# cv2.waitKey(0)
showImage(rotated)
```

# Smoothing an image

```python
# apply a Gaussian blur with a 11x11 kernel to the image to smooth it,
# useful when reducing high frequency noise
blurred = cv2.GaussianBlur(image, (11, 11), 0)
# cv2.imshow("Blurred", blurred)
# cv2.waitKey(0)
showImage(blurred)
```

```python
# draw a 2px thick red rectangle surrounding the face
output = image.copy()
cv2.rectangle(output, (150, 70), (700, 720), (0, 0, 255), 2)
# cv2.imshow("Rectangle", output)
# cv2.waitKey(0)
showImage(output)
```

```python
# draw a blue 20px (filled in) circle on the image centered at
# x=300,y=150
output = image.copy()
cv2.circle(output, (300, 150), 20, (255, 0, 0), -1)
# cv2.imshow("Circle", output)
# cv2.waitKey(0)
showImage(output)
```

```python
# draw green text on the image
output = image.copy()
cv2.putText(output, "OpenCV + Jurassic Park!!!", (10, 25), 
	cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
# cv2.imshow("Text", output)
# cv2.waitKey(0)
showImage(output)
```

# Detect Face

```python
import cv2

# Load the cascade
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

# Read the input image
img = cv2.imread('truongan.jpeg')

# Convert into grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detect faces
faces = face_cascade.detectMultiScale(gray, 1.1, 4)

# Draw rectangle around the faces
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)

# Display the output
# cv2.imshow('img', img)
# cv2.waitKey()
showImage(img)
```
